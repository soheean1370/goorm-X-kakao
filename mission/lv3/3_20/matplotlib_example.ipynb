```python
import matplotlib.pyplot as plt
import numpy as np 
companies = ['A사', 'B사', 'C사','D사', 'E사', 'F사']
market_share = np.array([40,25,15,8,7,5])

# 10% 이상인 경우에만 퍼센트 표시하는 함수
def autopct_over_10(pct):
    return f'{pct:.1f}%' if pct >= 10 else ''

plt.rc('font', family='Malgun Gothic')
plt.pie(market_share, labels=companies, autopct=autopct_over_10, 
        startangle=90, explode=[0.1, 0, 0, 0, 0, 0], 
        colors=['red','orange','yellow', 'green', 'blue', 'gray'])
plt.title('IT기업 시장 점유율')
plt.show()
```

# 제품 판매랑 비교 (Bar Chart)

```python
products = ["노트북", "태블릿", "스마트폰", "스마트워치", "헤드폰"]
sales = [150, 80, 300, 120, 180]
```

```python
plt.figure(figsize=(10, 6))  
plt.bar(products, sales, color='green', edgecolor='black')
plt.title('전자제품 판매량 비교', pad=15)  # 제목과 그래프 사이 여백 추가
plt.xlabel('제품명')
plt.ylabel('판매량')
plt.tight_layout()  # 자동으로 레이아웃 조정
plt.show()
```
