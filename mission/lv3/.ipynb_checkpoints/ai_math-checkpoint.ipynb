# 조건부 확률 계산

```python
import numpy as np

n= 10 # 총 사건 수
```

```python
# 0 : 발생하지 않음, 1: 발생함
a = np.random.binomial(1, 0.5 , n) # a가 발생할 확률
b = np.random.binomial(1, 0.5 , n) # b가 발생할 확률

print('A가 발생할 확률:', a)
print('B가 발생할 확률:',b)
```

```python
a_and_b = a*b # a와 b가 동시에 발생할 확률
print('A와B가 동시에 발생할 확률:',a_and_b)
```

```python
P_A = np.mean(a)
P_B = np.mean(b)
P_A_and_B = np.mean(a_and_b)
```

```python
# P(B|A): 사건 A가 발생한 조건에서 사건 B가 발생할 확률
P_B_given_A = P_A_and_B / P_A if P_A > 0 else 0

print('사건 A가 발생한 조건에서 사건 B가 발생할 확률:',P_B_given_A)
```

# 기본적인 확률 계산

```python
from scipy.stats import bernoulli
import numpy as np
import matplotlib.pyplot as plt

# 베르누이 확률분포 
p = 0.4  # 성공할 확률
q = 1 - p  # 실패할 확률
E, V = bernoulli.stats(p)
print(f'베르누이 분포의 기대값: {E:.3f}, 분산 {V:.3f}')

# 확률질량함수 시각화
plt.figure(figsize=(8, 5))
x = np.array([0, 1])
pmf = np.array([q, p])
plt.bar(x, pmf, color=['skyblue', 'salmon'], width=0.4)
plt.xticks([0, 1], ['실패(0)', '성공(1)'])
plt.ylabel('확률')
plt.title(f'베르누이 분포')
plt.grid(alpha=0.3)
plt.show()
```

```python
from scipy.stats import norm, uniform
import numpy as np
import matplotlib.pyplot as plt

# 균일분포 파라미터
a = 5  # 확률밀도함수 범위 시작
b = 15  # 확률밀도함수 범위 끝
x1 = 12  # 구할 범위 시작
x2 = 15  # 구할 범위 끝
p = 1/(b-a)  # 확률밀도함수 값
cdf = (x2-x1)*p  # 누적분포함수 차이
E = (a+b)/2  # 기대값
V = (b-a)**2 / 12  # 분산

print('12와 15그람 사이에서 가져갈 확률: {:.2f}'.format(cdf))
print(f'균일분포의 기대값: {E:.2f}, 분산: {V:.2f}')

# 시각화 코드
plt.figure(figsize=(10, 6))

# 균일분포 생성
rv = uniform(loc=a, scale=b-a)
x = np.linspace(a-1, b+1, 1000)
y = rv.pdf(x)

# 전체 분포 그리기
plt.plot(x, y, 'b-', lw=2, label='균일분포 PDF')

# 관심 영역(12-15) 색칠하기
x_fill = np.linspace(x1, x2, 100)
y_fill = rv.pdf(x_fill)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.3, label='12-15 구간')

# 그래프 설정
plt.xlim(a-1, b+1)
plt.ylim(0, p*1.2)
plt.xlabel('값 (그램)')
plt.ylabel('확률 밀도')
plt.title('균일분포 U(5, 15)와 12-15 구간의 확률')
plt.text(7, p/2, f'P(12≤X≤15) = {cdf:.2f}', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# 선형 방정식 계산

```python
import numpy as np

mat1 = np.array([[2 , -1 , 1], [3, 9 , -3], [1, 2, -3]])
mat2 = np.array([3, 12 ,-4])

x = np.linalg.solve(mat1, mat2)

for i in range(len(mat1)):
    equation = ' + '.join([f'{mat1[i][j]}*x{j+1}' for j in range(len(mat1[i]))])
    print(f'{equation} = {mat2[i]}')

print('x1 = {}, x2 = {}, x3 = {}'.format(x[0],x[1],x[2]))
print(np.allclose(np.dot(mat1, x), mat2)) # 검산

```
