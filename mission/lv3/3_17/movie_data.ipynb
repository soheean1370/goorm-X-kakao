```python
import pandas as pd

data = {
    "Title": [
        "Inception", "Interstellar", "The Dark Knight", "Parasite", "Avengers: Endgame",
        "The Matrix", "Gladiator", "Titanic", "Joker", "Mad Max: Fury Road",
        "The Godfather", "Pulp Fiction", "Schindler's List", "The Shawshank Redemption", "Fight Club"
    ],
    "Genre": [
        "Sci-Fi", "Sci-Fi", "Action", "Drama", "Action",
        "Sci-Fi", "Historical", "Romance", "Drama", "Action",
        "Crime", "Crime", "Drama", "Drama", "Drama"
    ],
    "Year": [
        2010, 2014, 2008, 2019, 2019,
        1999, 2000, 1997, 2019, 2015,
        1972, 1994, 1993, 1994, 1999
    ],
    "Director": [
        "Christopher Nolan", "Christopher Nolan", "Christopher Nolan", "Bong Joon-ho", "Anthony Russo",
        "Lana Wachowski", "Ridley Scott", "James Cameron", "Todd Phillips", "George Miller",
        "Francis Ford Coppola", "Quentin Tarantino", "Steven Spielberg", "Frank Darabont", "David Fincher"
    ],
    "Rating": [
        8.8, 8.6, 9.0, 8.6, 8.4,
        8.7, 8.5, 7.8, 8.5, 8.1,
        9.2, 8.9, 8.9, 9.3, 8.8
    ],
    "Audience_Million": [
        100, 120, 150, 80, 200,
        90, 50, 220, 85, 75,
        110, 95, 85, 130, 70
    ],
    "Budget_Million": [
        160, 165, 185, 11, 356,
        63, 103, 200, 55, 150,
        6, 8, 22, 25, 63
    ]
}

df = pd.DataFrame(data) 
```

```python
# 영화 장르가 Action인 영화만 출력
df.loc[df["Genre"]=="Action"]
```

```python
#2010년 이후 개봉한 영화만 출력
df.loc[df["Year"]> 2010]
```

```python
# 평점이 9.0 이상인 영화만 출력
df.loc[df["Rating"]>= 9.0]
```

```python
# Christopher Nolan이 감독한 영화만 출력
df.loc[df["Director"] == "Christopher Nolan"]
```

```python
# 제작비가 50백만 ~ 200백만 사이인 영화만 출력
df.loc[(df["Budget_Million"] > 50) & (df["Budget_Million"] < 200)]
```

```python
# 제작비가 100백만 이상인 영화만 필터링
df.loc[df["Budget_Million"] >= 100]
```

```python
# 관객수가 100백만 이상인 영화만 찾기
df.loc[df["Audience_Million"] >= 100]
```

```python
# 연도가 가장 큰값에 해당하는 (가장 최근) 해당하는 여화 찾기
df.sort_values(by="Year", ascending=False).head(1)
```

```python
# 가장 오래된 영화 찾기
df.sort_values(by="Year", ascending=True).head(1)
```
