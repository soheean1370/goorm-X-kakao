```python
import pandas as pd

data = pd.read_csv("modified_sales_data.csv")
```

```python
# 결측치가 포함된 행 보기
data[data.isnull().any(axis=1)]
```

# 결측치 처리

```python
# sales_amount 컬럼의 결측치를 평균값으로 채우기
data['sales_amount'].fillna(data['sales_amount'].mean(), inplace=True)

# discount_applied 컬럼의 결측치를 중앙값(median)으로 채우기
data['discount_applied'].fillna(data['discount_applied'].median(), inplace=True)

# payment_method 컬럼의 결측치를 최빈값으로(mode)으로 채우기
data['payment_method'].fillna(data['payment_method'].mode()[0], inplace=True)

#order_date의 결측치를 최근 날짜(max())로 채우기
data['order_date'].fillna(data['order_date'].max(), inplace=True)
data
```

```python
print("결측값 처리 후 데이터:")
print(data.isnull().sum())
```
