# 머신러닝 개념 초급 문제 세트
Python으로 배우는 인공지능을 위한 머신러닝 개념잡기 과목 문제입니다.

## 문제 1. 지도학습에서 레이블이 필요한 이유
다음 중 **지도학습에서 "정답 레이블(label)"이 필요한 이유**로 가장 적절한 설명은?

1. 데이터를 시각적으로 보기 쉽게 정렬하기 위해  
2. 입력 데이터의 차원을 줄이기 위해  
3. 모델이 입력과 출력 간의 관계를 학습할 수 있도록 하기 위해  
4. 분류(class) 이름을 예쁘게 출력하기 위해  

👉 정답: (3)  
👉 간단한 이유: 지도학습은 입력과 정답 레이블이 쌍으로 주어진 데이터를 바탕으로 학습한다. 모델이 이 쌍들을 통해 입력과 출력 간의 관계를 일반화하여 ㅇ예측할 수 있도록 한다

## 문제 2. 과적합(Overfitting)
다음 중 **과적합**이 발생한 모델의 특징으로 가장 적절한 것은 무엇인가?

1. 학습 데이터에서 성능이 낮고, 테스트 데이터에서 성능이 높다  
2. 학습 데이터에서 성능이 높고, 테스트 데이터에서 성능이 낮다  
3. 학습/테스트 모두에서 성능이 낮다  
4. 학습 데이터 없이도 예측이 가능하다  

👉 정답: (2)  
👉 간단한 이유: 모델이 훈련 데이터에 너무 과하게 맞춰져서, 테스트 데이터(새로운 데이터)에 대한 일반화 성능이 떨어지는 현상이 과적합이다

## 문제 3. KNN 알고리즘
KNN 알고리즘의 **예측 대상 데이터 포인트**가 있을 때, 이웃 중 3개가 “사과”, 2개가 “배”라고 한다면 예측 결과는?

1. 사과  
2. 배  
3. 예측 불가  
4. 평균값  

👉 정답: (1)  
👉 간단한 이유: knn 알고리즘은 가장 가까운 k개의 이웃 중 다수결로 예측하므로, 사과가 더 많으면 사과로 예측한다

## 문제 4. 거리 계산 방식
KNN 알고리즘에서 주로 사용되는 거리 측정 방식 중, 다음은 무엇을 의미하는가?

$$
\text{distance}(A, B) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2}
$$

1. 맨해튼 거리  
2. 유클리드 거리  
3. 자카드 유사도  
4. 코사인 유사도  

👉 정답: (2 )  
👉 간단한 이유: 주어진 식은 두 점 사이의 유클리드 거리 공식을 의미한다. 두 점 사이의 거리를 계산하는 공식이다

## 문제 5. 회귀와 분류의 차이
다음 중 **회귀(Regression)** 문제의 예로 가장 적절한 것은?

1. 이메일이 스팸인지 아닌지 판별  
2. 환자의 당뇨병 여부 분류  
3. 주택의 가격 예측  
4. 손글씨 숫자 판별  

👉 정답: (3)  
👉 간단한 이유: 회귀 문제는 연속적인 수치 값을 예측하는 문제이다. 

## 문제 6. k 값의 영향
KNN 알고리즘에서 **k 값을 너무 작게** 설정하면 생길 수 있는 현상은?

1. 과소적합  
2. 계산량 증가  
3. 이상치에 민감해짐  
4. 모델이 무작위로 예측함  

👉 정답: (3)  
👉 간단한 이유: k가 작으면 이상치 한개에도 영향을 많이 받아 불안정해진다

## 문제 7. KNN과 데이터 스케일링
두 특성(feature)을 가진 데이터셋에서, 하나는 0~1 범위의 값(예: 키), 다른 하나는 0~10,000 범위의 값(예: 연봉)일 때, **KNN 알고리즘을 사용할 경우 생길 수 있는 문제점**으로 가장 적절한 것은?

1. 연봉의 영향이 너무 작아져 예측 정확도가 떨어진다  
2. 키의 영향이 너무 커져 예측 정확도가 떨어진다  
3. 연봉의 값이 너무 커서 거리 계산에서 **지나치게 큰 영향**을 미친다  
4. 특성 수가 2개이므로 문제 없음  

👉 정답: (3)     
👉 간단한 이유: knn은 거리 기반 알고리즘이라 스케일이 큰 특성은 거리 계산에 과도한 영향을 미쳐 예측이 왜곡될 수 있다

## 문제 8. train_test_split() 함수의 목적
다음 중 `train_test_split()` 함수를 사용하는 **가장 주요한 목적**은 무엇인가?

1. 데이터를 시각화하기 위해 데이터를 나누는 함수  
2. 전체 데이터를 더 빠르게 처리하기 위해 나누는 함수  
3. 과적합을 방지하기 위해 학습용 데이터와 테스트용 데이터를 구분하는 함수  
4. 모델의 정확도를 높이기 위해 데이터를 랜덤하게 삭제하는 함수  

👉 정답: (3)  
👉 간단한 이유: 학습과 평가를 같은 데이터로 하면 과적합 위험이 커지므로, 데이터를 학습용과 테스트용으로 나누어 일반화 성능을 평가한다
