```python
from sympy import Limit, S, Symbol
x = Symbol('x')
limit_result  = Limit((x**2 - 4)/ (x-2), x, 2).doit()
print(limit_result)
```

```python
from sympy import symbols

def average(a,b):
    m = max(a,b)
    n = min(a,b)

    x = symbols('x')
    fx = x**2
    fb = fx.subs(x, m)
    fa = fx.subs(x, n)

    result = (fb - fa)/(b-a)
    return result

print(average(1,3))
```
