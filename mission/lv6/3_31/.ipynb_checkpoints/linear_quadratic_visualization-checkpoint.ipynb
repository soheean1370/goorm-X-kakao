# 함수 \( f(x) = 2x + 3 \) 와 \( f(x) = x^2 - 6x + 8 \)

## (1) 꼭짓점의 좌표  
이차 함수 \( f(x) = x^2 - 6x + 8 \) 의 꼭짓점은  
\[ x = -b / 2a = -(-6) / (2*1) = 6 / 2 = 3 \]  
일 때이고,  
\[ y = 3^2 - 6(3) + 8 = 9 - 18 + 8 = -1 \]  
따라서 꼭짓점은 (3, -1)  

## (2) x축과 만나는 점 (x절편)  
- **선형 함수**: \( 2x + 3 = 0 \)  
  \[ x = -3 / 2 = -1.5 \]  
  x절편은 (-1.5, 0)  

- **이차 함수**: \( x^2 - 6x + 8 = 0 \)  
  \[ (x-2)(x-4) = 0 \-> x = 2, 4 \]  
  x절편은 (2,0), (4,0)  

## (3) y축과 만나는 점 (y절편)  
- **선형 함수**: \( x = 0 \) 일 때  
  \[ f(0) = 2(0) + 3 = 3 \]  
  y절편은 (0,3)  

- **이차 함수**: \( x = 0 \) 일 때  
  \[ f(0) = 0^2 - 6(0) + 8 = 8 \]  
  y절편은 (0,8)  


```python
import numpy as np
import matplotlib.pyplot as plt

# x값 범위 설정
x_vals = np.linspace(-4, 10, 100)

# 함수 정의
y_linear = 2 * x_vals + 3
y_quadratic = x_vals**2 - 6*x_vals + 8

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)

# 선형 함수
plt.plot(x_vals, y_linear, label='y = 2x + 3', color='blue')

# 이차 함수
plt.plot(x_vals, y_quadratic, label='y = x^2 - 6x + 8', color='red')

# 절편 및 꼭짓점 표시
plt.scatter([-1.5, 2, 4, 0, 0, 3], [0, 0, 0, 3, 8, -1], color='black', zorder=3)
plt.text(-1.5, 0, '(-1.5, 0)', verticalalignment='bottom', horizontalalignment='right')
plt.text(2, 0, '(2, 0)', verticalalignment='bottom', horizontalalignment='right')
plt.text(4, 0, '(4, 0)', verticalalignment='bottom', horizontalalignment='right')
plt.text(0, 3, '(0, 3)', verticalalignment='bottom', horizontalalignment='left')
plt.text(0, 8, '(0, 8)', verticalalignment='bottom', horizontalalignment='left')
plt.text(3, -1, '(3, -1)', verticalalignment='top', horizontalalignment='left')

# 제목 및 범례 추가
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.xlabel('x')
plt.ylabel('y')
plt.title('선형 및 2차 함수 그래프')
plt.legend()
plt.show()
```
