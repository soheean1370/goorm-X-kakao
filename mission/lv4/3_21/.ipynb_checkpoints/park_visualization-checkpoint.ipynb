## Pandas로 데이터 불러오기

```python
import pandas as pd
import numpy as np

data = pd.read_csv('seoul_parks.csv')
data
```

## matplotlib 시각화

```python
import folium

# 지도 생성 (초기 위치 설정은 데이터의 첫 번째 공원의 위도와 경도)
m = folium.Map(location=[data['위도'].mean(), data['경도'].mean()], zoom_start=12, width=800, height=500)

# 공원의 각 위치에 마커 추가
for idx, row in data.iterrows():
    # 마커 생성, 팝업 내용은 공원이름과 면적
    folium.Marker(
        location=[row['위도'], row['경도']],  # 공원의 위도, 경도
        popup=f"<strong>공원 이름:</strong> {row['공원이름']}<br><strong>면적:</strong> {row['면적']}㎡",
        icon=folium.Icon(color='blue', icon='info-sign')  # 마커 색상 설정
    ).add_to(m)

# 지도 출력
m
```

##  seaborn 시각화 

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
plt.rc('font', family='Malgun Gothic')

sns.boxplot(x=data['구'], y=data['면적'], hue=None, palette="pastel", showfliers=False)

# y축 로그 스케일 적용
plt.yscale('log')

# 축 및 제목 설정
plt.xlabel('구')
plt.ylabel('공원 면적 (log scale)')
plt.title('구별 공원 면적 분포')

plt.xticks(rotation=45)
plt.show()
```

## plotly 시각화 

```python
import plotly.express as px

# Plotly 산점도 생성
fig = px.scatter(
    data,
    x="면적",  # 공원 면적
    y="구",   # 구 이름
    size="면적",  # 공원 면적
    hover_name="이름",  # 시 공원 이름 표시
    title="구별 공원 면적 분포",
    color="구",  # 구별
)

fig.show()
```
