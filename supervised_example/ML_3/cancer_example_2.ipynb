```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

os.chdir("../pydata")

df = pd.read_csv('cancer.csv')
print(df.head())
print("*******************")
print(df.shape)
```

```python
# 데이터 확인
X = df.iloc[:, 0:30]
y = df.iloc[:,30]

print(X.head())
print(y.head())
```

```python
# 모델 생성,훈련
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
print(X_train.head())
print(y_train.head())

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X,y)
```

```python
# 정확도 확인
print("prediction:",gnb.predict(X_test))
print("훈련 데이터 점수:", gnb.score(X_train, y_train))
print("테스트 데이터 점수:", round(gnb.score(X_test,y_test), 3))
```
