# Logistic Regression

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
os.chdir("../pydata")

df = pd.read_csv('cancer.csv')
print(df.head())
print("***************")
print(df.shape)
```

```python
# 데이터 확인
X = df.iloc[:, 0:30]
y = df.iloc[:, 30]

print(X.head())
print("***************")
print(y.head())
print("***************")
print(X.shape)
print(y.shape)
```

```python
# 타겟 컬럼 계층화
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
print(X_train.head())
print("***************")
print(y_train.head())
```

```python
# 모델 생성, 훈련
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# 사용 알고리즘 명시
logreg = LogisticRegression(solver="liblinear")
logreg.fit(X_train, y_train)
```

```python
# 정확도 확인
print("prediction:", logreg.predict(X_test))
print("훈련 데이터 점수:", logreg.score(X_train, y_train))
print("테스트 데이터 점수:", logreg.score(X_test, y_test))
print("테스트 데이터 점수(반올림):", round(logreg.score(X_test, y_test), 3))
```

```python
# 규제 낮추기
logreg = LogisticRegression(C=100, solver="liblinear")
logreg.fit(X_train, y_train)

print("훈련 데이터 점수:", logreg.score(X_train, y_train))
print("테스트 데이터 점수:", logreg.score(X_test, y_test))
print("테스트 데이터 점수(반올림):", round(logreg.score(X_test, y_test), 3))
```

```python
# 규제 변경
logreg = LogisticRegression(C=100, solver="liblinear", penalty="l1")
logreg.fit(X_train, y_train)

print("prediction:", logreg.predict(X_test))
print("훈련 데이터 점수:", logreg.score(X_train, y_train))
print("테스트 데이터 점수:", logreg.score(X_test, y_test))
print("테스트 데이터 점수(반올림):", round(logreg.score(X_test, y_test), 3))
```

```python
# 최대 훈련 횟수 변경
logreg = LogisticRegression(C=100, solver="liblinear", penalty="l1", max_iter=150)
logreg.fit(X_train, y_train)

print("prediction:", logreg.predict(X_test))
print("훈련 데이터 점수:", logreg.score(X_train, y_train))
print("테스트 데이터 점수:", logreg.score(X_test, y_test))
print("테스트 데이터 점수(반올림):", round(logreg.score(X_test, y_test), 3))
```
